```python
#segundo punto de la actividad, escogimos seleccion de bordes
# Librerias necesarias
from skimage import io 
from skimage import filters 
from skimage.color import rgb2gray
import matplotlib.pyplot as plt

# Abrimos la imagen
imagen = io.imread("imagen_de_monedas_griegas.jpg") # la ruta de la imagen
imagen_g = rgb2gray(imagen) # color de la imagen

# Filtros de la imagen: sobel, roberts, prewitt
filtros = [filters.sobel, filters.roberts, filters.prewitt]

for filtro in filtros: # este for se utiliza el iterador filtros y se recorre la lista filtros
    # en esta linea de codigo se aplica los filtos a traves de una lista
    img_fill = filtro(imagen_g)
    
    plt.imshow(img_fill) # se muestra el resultado y como es una funcion
    plt.show() # en esta linea de codigo da a entender al programa que muestre como tal la imagen
```

```python
import random
import numpy as np
from skimage import data
import matplotlib.pyplot as plt

ancho=200 # declarar la variable
imagen = [] # declarar una lista vacia
contador = 0 # declarar una variable contador
for i in range(0, ancho): # crear un for que va iterar de 0 a 200
  fila = [] # esta linea sirve para crear las filas de la matriz 
  for j in range(0, ancho): # con este for me sirve para iterar dentro de la matriz
    rojo = random.randint(0,255) # me da valores aleatorios de la intensidad entre 0 y 255
    pixel = (rojo)#ejemplo usando tupla
    if i < j: # en algebra lineal cuando I es menor que J estas en la diagonal de arriba
      pixel = 255 # pixel = 255 significa que esta en la maxima intensidad es decir color blanco
      fila.append(pixel) # en todas las filas de la matriz me va inserta el color blanco arriba de la diagonal principal
    else: # sino
       pixel = 0 # pixel = 0 significa que esta en la minima intensidad es decir color negro
       fila.append(pixel) # en todas las filas de la matriz me va insertar el color negro abajo de la diagonal principal
    contador +=1 # este contador significa la cantidad de pixeles que va haber en la matriz 
  imagen.append(fila) # en la lista me va insertar la cantidad de filas que hay en la matriz y por tanto la cantidad de columnas al estar en el segundo for ya que este segundo for recorre toda la matriz en orden fila-columna
imagen_aleatoria=imagen # en esta variable me va guardar como tal la matriz
plt.imshow(imagen_aleatoria) # va mostrar como tal la matriz
print(imagen_aleatoria) # en esta linea de codigo la va a imprimir  
print('contador queda con : '+str(contador)) # me va decir la cantidad de pixeles que hay en la matriz
```
